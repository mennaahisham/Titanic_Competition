According to number of null values to each feature we have to decide whether to fill or delete these null values and the way to fill it

```python
#we can see 4 features have null values: age, cabin, embarked
print("cabin null values=",train.Cabin.isnull().sum() / len(train))
print("age null values=",train.Age.isnull().sum() / len(train))
print("Embarked null values=",train.Embarked.isnull().sum() / len(train))
```

```python
#cabin feature 80% of data is null so we will drop
train.drop(columns=['Cabin'] , inplace=True, axis=1)
# drop cabin feature from test too
test.drop(columns=['Cabin'] , inplace=True, axis=1)
train.isnull().sum()
```

```python
#age feature will be replaced using average age values
train['Age'] = train['Age'].fillna(train['Age'].mean())
test['Age'] = test['Age'].fillna(test['Age'].mean())
train.isnull().sum()
```

```python
train['Embarked'].value_counts()
```

```python
#replace 2 null values using mod (the most repeated)
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
train.isnull().sum()
```

#fare in test file has one fare null value will be replaced by mean

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].mode()[0]) 
```

to change all values to numeric

```python
train.head()
train.drop(columns=['Name','Ticket','Died'] , inplace=True, axis=1)
test.drop(columns=['Name','Ticket'] , inplace=True, axis=1)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

in the previous check we made sure all our cleaning is done correctly and null values are completely disappear

```python
# mapping to numrical values
train['Sex'] = train['Sex'].map({'male':1, 'female':0})
train['Embarked'] = train['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
train.head(10)
```

```python
# mapping to numrical values
test['Sex'] = test['Sex'].map({'male':1, 'female':0})
test['Embarked'] = test['Embarked'].map({'S':0, 'C':1, 'Q':2})
```

```python
test.head()
```

```python
test.shape
```

# classification model 

prepare data

```python
y = train['Survived']
X = train.drop('Survived', axis=1)

Xtest = test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",clf.score(X_test, y_test))
```

cross validation 

```python
# scores = cross_val_score(clf, X, y, cv=5)
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

scores = cross_val_score(clf, X, y, cv=k_fold, n_jobs=1, scoring='accuracy')
score =round(np.mean(scores)*100, 2)
print("cross validation score=", score)
```

```python
predictions = clf.predict(Xtest)
```

```python
# output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output = pd.DataFrame(predictions, columns = ['Survived'])
output.set_index(test.index, inplace = True)
```

```python
output['Survived']  = output['Survived'].astype(int)
output.head(10)
```

here we predicted the rest of passengers whether they've survived or not, using random forest classifier
we evaluated the model using cross validation
